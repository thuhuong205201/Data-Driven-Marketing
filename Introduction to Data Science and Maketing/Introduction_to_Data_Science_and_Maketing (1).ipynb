# Indentify dissatisfied Employees

```python
def update_vals(value):
    if pd.isnull(value):
        return np.nan
    elif value == '-':
        return False
    else:
        return True
        
tafe_resignations['dissatisfied'] = tafe_resignations[['Contributing Factors. Dissatisfaction', 'Contributing Factors. Job Dissatisfaction']].applymap(update_vals).any(1, skipna = False)
tafe_resignations_up = tafe_resignations.copy()

tafe_resignations_up['dissatisfied'].value_counts(dropna = False)
```

```python
dete_resignations['dissatisfied'] = dete_resignations[['job_dissatisfaction', 'dissatisfaction_with_the_department', 'physical_work_environment', 'lack_of_recognition', 'lack_of_job_security', 'work_location', 'employment_conditions', 'work_life_balance', 'workload']].any(1, skipna=False)
dete_resignations_up = dete_resignations.copy()

dete_resignations_up['dissatisfied'].value_counts(dropna = False)
```

# Combine

```python
dete_resignations_up['institute'] = 'DETE'
tafe_resignations_up['institute'] = 'TAFE'
```

```python
combined = pd.concat([dete_resignations_up, tafe_resignations_up], ignore_index = True)

# Verify the number of non null values in each column
combined.notnull().sum().sort_values()
```

```python
combined.head()
```

# Clean the Service column

```python
combined = combined.rename(columns={
    'lengthofserviceoverall._overall_length_of_service_at_institute_(in_years)': 
                 'institute_service'})
```

```python
combined['institute_service'].value_counts()
```

```python
extracted_years = (combined['institute_service']
                  .astype(str).str.extract(r'(\d+)')
                  .astype('float'))

combined['institute_service_up'] = extracted_years

combined['institute_service_up'].value_counts()
```

```python
def career_stage(element):
    if element < 3:
        return 'New'
    elif 3 <= element < 7:
        return 'Experienced'
    elif 7 <= element < 11:
        return 'Established'
    elif pd.isnull(element):
        return np.nan
    else:
        return 'Veteran'

combined['service_cat'] = combined['institute_service_up'].apply(career_stage)

combined['service_cat'].value_counts()
```
